## TESTE T PARA DUAS MÉDIAS
## Amostras pareadas (dependentes)

### Exercício 01

```python
x = np.array([8.1, 7.9, 6.8, 7.8, 7.6, 7.9, 5.7, 8.4, 8.0, 9.5, 8.0, 6.8])
y = np.array([11.6, 8.8, 9.9, 9.5, 11.6, 9.1, 10.6, 10.8, 13.4, 10.6, 10.5, 11.4])
di = x - y
print(x)
print
print(y)
print(di)
```

```python
n = len(di)
di_med = di.mean()
di_std = di.std(ddof = 1)
print("n: {:6} Média di: {:10.4f} Desvio Padrão di: {:10.4f}".format(n, di_med, di_std))
```

```python
tp = hip.media_testet_amostras_dependentes(di_med, di_std, n)
tp
```
